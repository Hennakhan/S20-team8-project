```python
import tensorflow as tf
from PIL import Image
import matplotlib.pyplot as plt
import os
import numpy as np 
```

```python
# loading model 
model = tf.keras.models.load_model("food_model3.h5")
```

```python
# 
CATEGORIES = ["Apple pie", "Waffle", "Churro"]

# change the file name if you want to test it with different picture
image_path="input/churro.JPEG"

files = os.listdir("input")
files[:10]
```

```python

#im = Image.open("image_path")
#im.convert('RGB').save("image_name.jpg","JPEG")
#im = Image.open("image_path").resize(x,y)

img_resize = (224,224)
apple_train = []
test=[]

img = Image.open(image_path)
img = img.resize(img_resize)
test.append(img)
img = np.asarray(img)
apple_train.append(img)


data_train = np.array(apple_train, dtype=np.float32)
data_train /= 255
```

```python
prediction = model.predict([data_train])
print(prediction)  # will be a list in a list.
result = CATEGORIES[np.argmax(prediction)]
print(result)
```
